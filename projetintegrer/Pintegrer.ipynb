```python
df.columns = [
    'timestamp',
    'username',
    'age',
    'gender',
    'education_level',
    'academic_performance_rating',
    'extracurricular_involvement',
    'participating_activities',
    'time_management_strategy',
    'impact_on_time_management',
    'institutional_support',
    'institutional_encouragement_rating',
    'extracurricular_contribution_rating',
    'improved_peer_relations',
    'influence_on_personal_development',
    'preferred_activities',
    'frequency_of_participation',
    'weekly_hours_spent_on_activities',
    'current_overall_average',
    'participation_in_events_last_year',
    'satisfaction_rating',
    'skills_acquired_count'
]

df.head()
```

```python
# Extract only numeric values from the "age" column
df['age'] = df['age'].astype(str).str.extract('(\d+)')

# Convert the "age" column to numeric type
df['age'] = pd.to_numeric(df['age'])

# Print the updated DataFrame
print(df)
```

```python
df
```

```python
# Extract only numeric values from the "weekly_hours_spent_on_activities" column
df['weekly_hours_spent_on_activities'] = df['weekly_hours_spent_on_activities'].astype(str).str.extract('(\d+)')

# Convert the "weekly_hours_spent_on_activities" column to numeric type
df['weekly_hours_spent_on_activities'] = pd.to_numeric(df['weekly_hours_spent_on_activities'])

# Print the updated DataFrame
print(df)
```

```python
df
```

```python
# Replace values in the "participation_in_events_last_year" column
df['participation_in_events_last_year'] = df['participation_in_events_last_year'].replace({
    'Pas encore': 0,
    'Une seule fois': 1,
    '2 fois par semaine et parfois plus': 2
})

#df Extract only numeric values from the "participation_in_events_last_year" column
df['participation_in_events_last_year'] = pd.to_numeric(df['participation_in_events_last_year'], errors='coerce')

# Print the updated DataFrame
print(df)
```

```python
df
```

```python
# Remove "ème année" and "ère année" from the "education_level" column
df['education_level'] = df['education_level'].str.replace(r'ème année|ère année', '', regex=True)

# Print the updated DataFrame
print(df['education_level'])
```

```python
df
```

```python
df = df.fillna(0)
```

```python
df
```

```python
df['participating_activities'] = df['participating_activities'].fillna(0)
df
```

```python
# Bar plot for gender distribution
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()

# Bar plot for education level distribution
sns.countplot(x='education_level', data=df)
plt.title('Education Level Distribution')
plt.show()
```

```python
# Box plot for age distribution
sns.boxplot(x='age', data=df)
plt.title('Age Distribution')
plt.show()

# Box plot for academic performance rating
sns.boxplot(x='academic_performance_rating', data=df)
plt.title('Academic Performance Rating Distribution')
plt.show()

```

```python
# Scatter plot for age vs. weekly hours spent on activities
sns.scatterplot(x='age', y='weekly_hours_spent_on_activities', data=df)
plt.title('Age vs. Weekly Hours Spent on Activities')
plt.show()

```

```python
# Pie chart for participation frequency
df['frequency_of_participation'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Frequency of Participation')
plt.show()
```

```python
# Grouped bar plot for gender and education level
sns.countplot(x='education_level', hue='gender', data=df)
plt.title('Gender Distribution by Education Level')
plt.show()
```

```python
# Grouped bar plot for gender and ACADEMIC PERFORMANCE
sns.countplot(x='academic_performance_rating', hue='gender', data=df)
plt.title('Gender Distribution by academic_performance_rating')
plt.show()
```

```python
# Grouped bar plot for time management strategy and gender
sns.countplot(x='time_management_strategy', hue='gender', data=df)
plt.title('Distribution of Time Management Strategies by Gender')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Get unique strategies and corresponding colors
strategies = df['time_management_strategy'].unique()
colors = sns.color_palette('Set2', n_colors=len(strategies))

# Plot pie chart without text
strategy_counts = df['time_management_strategy'].value_counts()
plt.pie(strategy_counts, labels=[''] * len(strategy_counts), autopct='', startangle=90, wedgeprops=dict(width=0.4))

# Create a legend with a table on the right
legend_labels = [f'{strategy} ({count}%)' for strategy, count in zip(strategy_counts.index, strategy_counts / strategy_counts.sum() * 100)]
legend_table = pd.DataFrame({'Label': legend_labels, 'Color': colors})

plt.legend(legend_labels, loc='center left', bbox_to_anchor=(1, 0.5), title='Time Management Strategy', frameon=False)
plt.axis('off')  # Remove unnecessary elements


plt.show()

```

```python
# Example using Chi-square test for independence
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['gender'], df['preferred_activities'])

# Perform Chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"Chi-square value: {chi2}, p-value: {p}")
```
